```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PK(x):
    return np.maximum(x, 0)

def gradient_projeté_quad(A, b, x0, rho = 0.01, eps = 1e-6, Nmax = 5000):
    x = np.array(PK(x0), dtype=float)
    n = 0
    LJ = [np.linalg.norm(A @ x - b)]
    LX = [x.copy()]
    while LJ[-1] > eps and n < Nmax:
        grad = A @ x - b
        x -= rho * grad
        x = PK(x)
        n += 1
        LJ.append(np.linalg.norm(A @ x - b))
        LX.append(x.copy())
    if n == Nmax:
        print("Attention : Convergence non atteinte après", Nmax, "itérations.")
    return x, n, LJ, np.array(LX)

def gradJ(x):
    return (2 * M.T @ (M @ x - f))

def gradient_projeté(gradJ, x0, rho = 0.01, eps = 1e-6, Nmax = 5000):
    x = np.array(PK(x0), dtype=float)
    n = 0
    LJ = [np.linalg.norm(gradJ(x0))]
    LX = [x.copy()]
    while LJ[-1] > eps and n < Nmax:
        grad = gradJ(x)
        x -= rho * grad
        x = PK(x)
        n += 1
        LJ.append(np.linalg.norm(gradJ(x)))
        LX.append(x.copy())
    if n == Nmax:
        print("Attention : Convergence non atteinte après", Nmax, "itérations.")
    return x, n, LJ, np.array(LX)

M = np.array([[1, 2],[-1, 3]])
f = np.array([-3,4])
x0 = np.array([8, -2])
A = 2 * M.T @ M
b = 2 * M.T @ f


x_proj_quad, n_proj_quad, LJ_proj_quad, LX_proj_quad = gradient_projeté_quad(A, b, x0)
x_proj, n_proj, LJ_proj, LX_proj = gradient_projeté(gradJ, x0)
print('LJ_proj_quad =',LJ_proj_quad[-5:-1])
print('LJ_proj =', LJ_proj[-5:-1])
print("Minimum (Gradient Projeté Quadratique) :", x_proj_quad, "| Itérations :", n_proj_quad)
print("Minimum (Gradient Projeté) :", x_proj, "| Itérations :", n_proj)

# ---- AFFICHAGE ----
plt.figure(figsize=(12, 5))

# Plot de la norme du gradient
plt.subplot(1, 2, 1)
plt.plot(LJ_proj_quad, marker='o', linestyle='-', color='b', label="Gradient Projeté Quadratique")
plt.plot(LJ_proj, marker='s', linestyle='-', color='r', label="Gradient Projeté")
plt.yscale("log")  
plt.xlabel("Itérations")
plt.ylabel("Norme du gradient") 
plt.title("Convergence de la norme du gradient")
plt.legend()
plt.grid()

# Trajectoire dans l'espace des solutions
plt.subplot(1, 2, 2)
plt.plot(LX_proj_quad[:, 0], LX_proj_quad[:, 1], marker='o', linestyle='-', color='b', label="Gradient Projeté Quadratique")
plt.plot(LX_proj[:, 0], LX_proj[:, 1], marker='s', linestyle='-', color='r', label="Gradient Projeté")
plt.scatter(LX_proj_quad[0, 0], LX_proj_quad[0, 1], color='g', s=100, label="Départ", zorder=3)
plt.scatter(LX_proj_quad[-1, 0], LX_proj_quad[-1, 1], color='b', s=100, label="Arrivée Gradient Projeté Quadratique", zorder=3)
plt.scatter(LX_proj[0, 0], LX_proj[0, 1], color='g', s=100, zorder=3)  # Départ aussi pour conjugué
plt.scatter(LX_proj[-1, 0], LX_proj[-1, 1], color='r', s=100, label="Arrivée Gradient Projeté", zorder=3)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Trajectoire de x dans l'espace")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
