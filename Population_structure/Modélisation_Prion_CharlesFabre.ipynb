## Modélisation de la dynamique du prion

L’encéphalopathie spongiforme, dont la forme humaine est la maladie de Creutzfeldt-Jakob, est causée par un agent infectieux appelé **prion**. Il s’agit d’une forme anormale de la protéine naturelle **PrPC** (cellulaire), notée **PrPSc** (scrapie), capable de provoquer la conversion des protéines saines en formes pathogènes. Cette conversion entraîne la formation d’**agrégats protéiques** toxiques et résistants aux traitements, perturbant les fonctions neuronales.

Dans le modèle présenté, on suit l’évolution temporelle de trois grandeurs biologiques :
- $V(t)$ : concentration des **monomères sains** PrPC,
- $U(t)$ : **nombre d’agrégats infectieux** (PrPSc),
- $P(t)$ : **masse totale** de protéines agglomérées.

Le comportement du système est régi par des phénomènes de **polymérisation**, **fragmentation**, et **dégradation**, conduisant à un système d'équations différentielles ordinaires simplifié :

$$
\begin{cases}
\frac{dU}{dt} = \bar{\beta} P - (\mu + 2\bar{\beta} x_0) U \\
\frac{dV}{dt} = \lambda - \gamma V - \tau VU + \bar{\beta} x_0^2 U \\
\frac{dP}{dt} = \tau VU - \mu P - \bar{\beta} x_0^2 U
\end{cases}
$$

Ce système modélise la compétition entre l'accumulation des agrégats et la réponse de l'organisme via la production et la dégradation des protéines.

---

## Théorème 1 — Existence et unicité des solutions

Nous allons démontré que pour toute condition initiale $(U_0, V_0, P_0)$ appartenant à l’ensemble

$$
X := \{ (a,b,c) \in \mathbb{R}^3 \mid a \geq 0,\ b \geq 0,\ c \geq x_0 a \},
$$

le système admet une **unique solution globale définie sur $\mathbb{R}_+$**, qui reste dans $X$ au cours du temps.

Ce résultat garantit la **cohérence physique** du modèle, et constitue le fondement de l’étude des comportements asymptotiques (états sain ou infecté).

---



## Théorème

**Énoncé** : Pour toute donnée initiale $(U_0, V_0, P_0) \in X := \{ (a,b,c) \in \mathbb{R}^3 \ | \ a \geq 0, \ b \geq 0, \ c \geq x_0 a \}$, le système suivant :

$$
\begin{cases}
\frac{dU}{dt} = \beta P - (\mu + 2\beta x_0)U \\
\frac{dV}{dt} = \lambda - \gamma V - \tau VU + \beta x_0^2 U \\
\frac{dP}{dt} = \tau VU - \mu P - \beta x_0^2 U
\end{cases}
$$

avec $\beta, \lambda, \gamma, \tau, x_0, \mu > 0$, admet une **unique solution** $(U(t), V(t), P(t))$ définie sur $\mathbb{R}_+$, telle que $(U(t),V(t),P(t)) \in X$ pour tout $t \geq 0$.

---

## Preuve

### 1. Existence locale et unicité

On définit la fonction $F : \mathbb{R}^3 \to \mathbb{R}^3$ par :

$$
F(U,V,P) = 
\begin{pmatrix}
\beta P - (\mu + 2\beta x_0)U \\
\lambda - \gamma V - \tau VU + \beta x_0^2 U \\
\tau VU - \mu P - \beta x_0^2 U
\end{pmatrix}
$$

La fonction $F$ est **localement lipschitzienne** car elle est polynomiale en $(U,V,P)$.  
Par le **théorème de Cauchy-Lipschitz**, pour toute condition initiale dans $\mathbb{R}^3$, il existe une **unique solution locale** définie sur un intervalle $[0, T_{\max})$, avec $T_{\max} > 0$.

---

### 2. Invariance de l'ensemble $X$

On montre que si la condition initiale est dans $X$, alors la solution reste dans $X$ pour tout $t \geq 0$.

#### a) Positivité de $U(t)$

On suppose par l'absurde que $U(t)$ devient négatif. Soit $t^* = \inf\{ t > 0 \ | \ U(t) < 0 \}$.  
Par continuité, $U(t^*) = 0$. Alors :

$$
\frac{dU}{dt}(t^*) = \beta P(t^*) \geq 0
$$

Donc $U$ ne peut pas devenir strictement négatif : **$U(t) \geq 0$** pour tout $t \geq 0$.

#### b) Positivité de $V(t)$

Si $V(t) = 0$ à un instant $t$, alors :

$$
\frac{dV}{dt}(t) = \lambda + \beta x_0^2 U(t) \geq 0
$$

Donc $V$ ne peut pas passer en négatif : **$V(t) \geq 0$** pour tout $t \geq 0$.

#### c) Inégalité $P(t) \geq x_0 U(t)$

Posons $\phi(t) = P(t) - x_0 U(t)$, on veut montrer que $\phi(t) \geq 0$.

Calculons la dérivée :

$$
\frac{d\phi}{dt} = \frac{dP}{dt} - x_0 \frac{dU}{dt} = (\tau VU - \mu P - \beta x_0^2 U) - x_0 (\beta P - (\mu + 2\beta x_0)U)
$$

Développement :

$$
\frac{d\phi}{dt} = -(\mu + \beta x_0)P + \tau VU + x_0(\mu + \beta x_0)U = -(\mu + \beta x_0)\phi + \tau VU
$$

Il s'agit d'une équation différentielle linéaire d'ordre 1 sur $\phi$, avec un terme source $\tau VU \geq 0$ et un terme de décroissance en $\phi$.

Par le **lemme de Grönwall**, si $\phi(0) \geq 0$, alors $\phi(t) \geq 0$ pour tout $t$ :  
**$P(t) \geq x_0 U(t)$**, donc $(U,V,P)(t) \in X$.

---

### 3. Existence globale

Les dérivées de $U$, $V$, et $P$ sont des polynômes de degré 2 en les variables, avec des termes de perte proportionnels : par exemple :

$$
\frac{dU}{dt} = \beta P - (\mu + 2\beta x_0)U
$$

Le terme $-\mu P$ dans l’équation de $P$ empêche que $P(t)$ explose, de même pour $U$ et $V$.

On peut alors **borner** les solutions sur tout intervalle fini $[0,T]$, ce qui empêche les explosions en temps fini.  
Donc $T_{\max} = +\infty$ : la solution est définie **globalement**.

---

### ✅ Conclusion

La solution $(U(t), V(t), P(t))$ du système est **unique, globale sur $\mathbb{R}_+$**, et vérifie $(U(t), V(t), P(t)) \in X$ pour tout $t \geq 0$.


## Simulation du système UVP

Pour illustrer le comportement dynamique du système, nous proposons une simulation numérique selon deux types de conditions initiales :
- un **état sain**, avec une absence initiale d’agrégats ($U = 0$, $P = 0$),
- un **état infecté**, avec une présence initiale d’agrégats ($U > 0$, $P > 0$).

Ces deux situations permettront d'observer l’évolution respective du système et de motiver l'étude du **comportement asymptotique** via le Théorème 2.

```python
import numpy as np
import matplotlib.pyplot as plt
# Paramètres du modèle
tau = 1       # Taux de polymérisation
x0 = 0        # Taille critique
beta = 0.5    # Taux de fragmentation
mu = 1        # Taux de dégradation des agrégats
lam = 2       # Source naturelle de monomères
gamma = 1.5   # Taux de dégradation des monomères

# Système différentiel UVP
def UVP(U, V, P):
    dU = beta * P - (mu + 2 * beta * x0) * U
    dV = lam - gamma * V - tau * V * U + beta * U * x0**2
    dP = tau * V * U - mu * P - beta * U * x0**2
    return np.array([dU, dV, dP])

# Méthode d'Euler explicite pour un système à 3 dimensions
def Euler3D(f, y0, t0, tf, dt):
    n = int((tf - t0) / dt)
    Y = np.zeros((n + 1, 3))  # Stockage des solutions U, V, P
    Y[0, :] = y0
    for i in range(n):
        Y[i + 1, :] = Y[i, :] + dt * f(Y[i, 0], Y[i, 1], Y[i, 2])
    return Y

# Paramètres temporels
dt = 0.0001
t0, tf_infecte, tf_sain = 0, 10, 4
time_infecte = np.linspace(t0, tf_infecte, int((tf_infecte - t0) / dt) + 1)
time_sain = np.linspace(t0, tf_sain, int((tf_sain - t0) / dt) + 1)

# Résolution pour l'état sain
y0_sain = [0, 0.2, 0]
result_sain = Euler3D(UVP, y0_sain, t0, tf_sain, dt)

# Résolution pour l'état infecté
y0_infecte = [1, 0.2, 1.5]
result_infecte = Euler3D(UVP, y0_infecte, t0, tf_infecte, dt)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# Tracé des résultats pour l'état sain
ax1.plot(time_sain, result_sain[:, 0], label='U (PrPSc)', color='blue')
ax1.plot(time_sain, result_sain[:, 1], label='V (PrPC)', color='green')
ax1.plot(time_sain, result_sain[:, 2], label='P (PrPSc)', color='red')
ax1.set_xlabel('Temps')
ax1.set_title("Évolution du système UVP - État sain")
ax1.legend()
ax1.grid()


# Tracé des résultats pour l'état infecté
ax2.plot(time_infecte, result_infecte[:, 0], label='U (PrPSc)', color='blue')
ax2.plot(time_infecte, result_infecte[:, 1], label='V (PrPC)', color='green')
ax2.plot(time_infecte, result_infecte[:, 2], label='P (PrPSc)', color='red')
ax2.set_xlabel('Temps')
ax2.set_title("Évolution du système UVP - État infecté")
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

## Théorème 2 — Comportement asymptotique du système

La simulation précédente met en évidence deux régimes dynamiques possibles selon les conditions initiales et les paramètres du modèle : soit le système converge vers un **état sain** sans agrégats, soit il évolue vers un **état infecté** avec une charge prion persistante.

Le **Théorème 2-(i)** précise rigoureusement les conditions dans lesquelles **l’état sain** est **globalement attractif**. Plus précisément :

> **Théorème 2-(i)**  
> Si la condition suivante est satisfaite :
>
> $$
> \sqrt{ \dfrac{ \bar{\beta} \lambda \tau }{ \gamma } } < x_0 \bar{\beta} + \mu,
> $$
> 
> alors l’état sain $(U, V, P) = (0, \lambda/\gamma, 0)$ est **globalement stable**, et pour toute condition initiale $(U_0, V_0, P_0) \in X$, la solution du système vérifie :
>
> $$
> (U(t), V(t), P(t)) \longrightarrow (0, \lambda/\gamma, 0) \quad \text{quand } t \to \infty.
> $$

Ce résultat garantit l’extinction des agrégats dans le temps, même en présence d’une faible quantité initiale, à condition que la fragmentation ou la polymérisation soient suffisamment faibles, ou que la dégradation soit dominante.

Nous allons à présent justifier ce théorème analytiquement.


### Justification du Théorème 2-(i) — Stabilité globale de l’état sain

On souhaite démontrer que, sous la condition :
$$
\sqrt{ \frac{ \bar{\beta} \lambda \tau }{ \gamma } } < x_0 \bar{\beta} + \mu,
$$
le système converge globalement vers l’état sain $(U, V, P) = (0, \lambda/\gamma, 0)$.

#### Idée de la preuve

On construit une fonction de type Lyapunov, qui mesure l’écart par rapport à l’état sain, et on montre que cette fonction décroît strictement le long des trajectoires du système.

#### Fonctionnelle candidate

Pour $A > 0$, on considère :
$$
\mathcal{L}(t) = \frac{1}{2} \left( V(t) - \frac{\lambda}{\gamma} \right)^2 + A P(t) + A \frac{\mu}{\bar{\beta}} U(t)
$$

Cette fonction est positive et s’annule uniquement à l’état sain. On calcule sa dérivée le long des solutions du système (4) :

$$
\frac{d\mathcal{L}}{dt} = -\gamma \left( V - \frac{\lambda}{\gamma} \right)^2 - U \, \Pi_A(V)
$$

où $\Pi_A(V)$ est un polynôme du second degré en $V$ donné par :
$$
\Pi_A(V) = A \left( \tau V - \bar{\beta} x_0^2 \right) - \frac{\mu}{\bar{\beta}} \left( \bar{\beta} V - (\mu + 2 \bar{\beta} x_0) \right)
$$

#### Choix de $A$ et signe de $\Pi_A(V)$

La condition sur les paramètres assure qu’il existe $A > 0$ et $\pi > 0$ tels que pour tout $V \in \mathbb{R}$ :
$$
\Pi_A(V) \geq \pi > 0
$$

Alors, on a :
$$
\frac{d\mathcal{L}}{dt} \leq -\gamma \left( V - \frac{\lambda}{\gamma} \right)^2 - \pi U(t) \leq 0
$$

avec égalité seulement à l’état sain. Par le **théorème de LaSalle**, on conclut que toute trajectoire converge vers l’unique point d’annulation de $\mathcal{L}$, à savoir :
$$
(U, V, P) = (0, \lambda/\gamma, 0)
$$

---

### ✅ Conclusion

Sous la condition $\sqrt{ \bar{\beta} \lambda \tau / \gamma } < x_0 \bar{\beta} + \mu$, **l’état sain est globalement asymptotiquement stable**, quelle que soit la condition initiale dans $X$.


### Vers le Théorème 2-(ii)

Nous venons de démontrer que, lorsque la condition $\sqrt{ \bar{\beta} \lambda \tau / \gamma } < x_0 \bar{\beta} + \mu$ est satisfaite, le système converge vers l’état sain.

Nous nous intéressons maintenant à la situation opposée, où cette condition est violée. Dans ce cas, on observe numériquement que le système converge vers un **état infecté** non trivial, ce que confirme le **Théorème 2-(ii)**.


## Théorème 2 (ii) — Stabilité globale de l'état infecté

**Énoncé :**  
Si $\sqrt{\bar{\beta} \lambda \tau / \gamma} > x_0 \bar{\beta} + \mu$, alors **l'état infecté** $(\bar{U}, \bar{V}, \bar{P})$ est **globalement asymptotiquement stable**.

---

### 1. Détermination du point d’équilibre infecté

On considère le système suivant :

- $\dfrac{dU}{dt} = \beta P - (\mu + 2\beta x_0) U$
- $\dfrac{dV}{dt} = \lambda - \gamma V - \tau VU + \beta x_0^2 U$
- $\dfrac{dP}{dt} = \tau VU - \mu P - \beta x_0^2 U$

On cherche les points d'équilibre $(\bar{U}, \bar{V}, \bar{P})$ satisfaisant :

- $0 = \beta \bar{P} - (\mu + 2\beta x_0) \bar{U}$
- $0 = \lambda - \gamma \bar{V} - \tau \bar{V} \bar{U} + \beta x_0^2 \bar{U}$
- $0 = \tau \bar{V} \bar{U} - \mu \bar{P} - \beta x_0^2 \bar{U}$

On en déduit :

- $\bar{P} = \dfrac{\mu + 2\beta x_0}{\beta} \bar{U}$
- Puis, en remplaçant dans la troisième équation, on peut démontrer qu’il existe **un unique point d’équilibre strictement positif** $(\bar{U}, \bar{V}, \bar{P})$ si la condition :

$$
\sqrt{\bar{\beta} \lambda \tau / \gamma} > x_0 \bar{\beta} + \mu
$$

est satisfaite. Cette condition correspond à la **survenue d’une invasion infectieuse**, analogue à un **nombre de reproduction effectif $R_{\text{eff}} > 1$**.

---

### 2. Linéarisation autour de $(\bar{U}, \bar{V}, \bar{P})$

On note $F(U, V, P)$ le champ vectoriel défini par le système. La matrice jacobienne $J$ est donnée par :

$$
J =
\begin{bmatrix}
-(\mu + 2\beta x_0) & 0 & \beta \\
-\tau V + \beta x_0^2 & -\gamma - \tau U & 0 \\
\tau V - \beta x_0^2 & \tau U & -\mu
\end{bmatrix}
$$

En l’évaluant en $(\bar{U}, \bar{V}, \bar{P})$, on obtient une matrice $J(\bar{U}, \bar{V}, \bar{P})$ **avec toutes ses entrées finies et les paramètres strictement positifs.**

On applique alors le **théorème de stabilité linéaire** :

> **Théorème (stabilité locale par linéarisation)**  
> Si toutes les valeurs propres de $J(\bar{U}, \bar{V}, \bar{P})$ ont une partie réelle strictement négative, alors l’équilibre $(\bar{U}, \bar{V}, \bar{P})$ est localement asymptotiquement stable.

Dans notre cas :
- $\bar{U}, \bar{V} > 0$, donc $J$ a une structure assurant que toutes les valeurs propres ont une partie réelle strictement négative (peut être confirmé numériquement ou via Routh-Hurwitz).

On conclut donc que l’état infecté est **localement stable**.

---

### 3. Stabilité globale

Pour conclure à la stabilité **globale**, on utilise les arguments suivants :

#### a) Invariance de l'ensemble $X$
Le système évolue dans un domaine invariant :

$$
X = \{ (U, V, P) \in \mathbb{R}^3 : U \ge 0, V \ge 0, P \ge x_0 U \}
$$

et les trajectoires y restent pour tout $t \ge 0$.

#### b) Unicité du point d’équilibre infecté
Sous la condition du théorème, le système admet un **unique point d’équilibre** dans $X$ avec $U, V, P > 0$.

#### c) Instabilité de l’équilibre trivial
Dans ce cas, le point $(0, \lambda/\gamma, 0)$ est instable, ce qui exclut toute convergence vers l’état sain.

#### d) Pas de cycles limites
Le système est de dimension 3 avec structure dissipative (toutes les trajectoires restent bornées), ce qui permet d’écarter l'existence de cycles limites par le **théorème de Poincaré-Bendixson généralisé** (valable ici grâce à l’invariance et la compacité des trajectoires).

---

### 🔚 Conclusion

Par **unicité de l’équilibre**, **stabilité locale**, **invariance**, et **absence d'autres attracteurs**, toute trajectoire dans $X$ converge vers $(\bar{U}, \bar{V}, \bar{P})$ :

> **Donc, $(\bar{U}, \bar{V}, \bar{P})$ est globalement asymptotiquement stable.** ✅



### Simulation du comportement infecté

Pour illustrer le Théorème 2, nous allons maintenant simuler deux cas de figure :

- Un **état initial infecté** dont les paramètres **satisfont la condition** $\sqrt{ \bar{\beta} \lambda \tau / \gamma } > x_0 \bar{\beta} + \mu$,  
  ce qui entraîne une convergence vers un **état infecté stable**.

- Un **autre état initial infecté** mais avec des paramètres qui **ne vérifient pas cette condition**,  
  ce qui conduit à un retour vers l’**état sain** $(0, \lambda/\gamma, 0)$.

Ces simulations permettront d'observer concrètement comment la condition du théorème détermine le comportement asymptotique du système.


```python
import numpy as np
import matplotlib.pyplot as plt

tau = 1.5
x0 = 0
beta = 0.5
mu = 1
lam = 2
gamma = 1

# Système UVP
def UVP(U, V, P):
    dU = beta * P - (mu + 2 * beta * x0) * U
    dV = lam - gamma * V - tau * V * U + beta * U * x0**2
    dP = tau * V * U - mu * P - beta * U * x0**2
    return np.array([dU, dV, dP])

# Méthode d’Euler
def Euler3D(f, y0, t0, tf, dt):
    n = int((tf - t0) / dt)
    Y = np.zeros((n + 1, 3))
    Y[0, :] = y0
    for i in range(n):
        Y[i + 1, :] = Y[i, :] + dt * f(Y[i, 0], Y[i, 1], Y[i, 2])
    return Y

# ============
# État infecté
# ============
t0, tf, dt = 0, 20, 0.0001
time = np.linspace(t0, tf, int((tf - t0) / dt) + 1)
y0 = [1.0, 0.2, 1.5]
result = Euler3D(UVP, y0, t0, tf, dt)

# ==============================
# État infecté – Condition violée
# ==============================
x0_bad = 1.0
beta_bad = 0.1
mu_bad = 1.0
lam_bad = 1.0
gamma_bad = 2.0

def UVP_bad(U, V, P):
    dU = beta_bad * P - (mu_bad + 2 * beta_bad * x0_bad) * U
    dV = lam_bad - gamma_bad * V - tau * V * U + beta_bad * U * x0_bad**2
    dP = tau * V * U - mu_bad * P - beta_bad * U * x0_bad**2
    return np.array([dU, dV, dP])

y0_bad = [1.0, 0.2, 1.5]
result_bad = Euler3D(UVP_bad, y0_bad, t0, tf, dt)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# Tracé des résultats pour l'état infecté global
ax1.plot(time, result[:, 0], label='U (PrPSc)', color='blue')
ax1.plot(time, result[:, 1], label='V (PrPC)', color='green')
ax1.plot(time, result[:, 2], label='P (PrPSc)', color='red')
ax1.set_xlabel('Temps')
ax1.set_title("Évolution du système UVP - État infecté globalement stable")
ax1.legend()
ax1.grid()


# Tracé des résultats pour l'état infecté local
ax2.plot(time, result_bad[:, 0], label='U (PrPSc)', color='blue')
ax2.plot(time, result_bad[:, 1], label='V (PrPC)', color='green')
ax2.plot(time, result_bad[:, 2], label='P (PrPSc)', color='red')
ax2.set_xlabel('Temps')
ax2.set_title("Évolution du système UVP - État infecté converge vers état sain")
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()

def check_infection_stability(tau, lam, gamma, x0, beta, mu):
    LHS = np.sqrt((beta * lam * tau) / gamma)
    RHS = x0 * beta + mu
    print("Vérification de la condition du théorème 2-(ii) :")
    print(f"√(β * λ * τ / γ) = {LHS:.4f}")
    print(f"x₀ * β + μ      = {RHS:.4f}")
    if LHS > RHS:
        print("✅ Condition satisfaite : l’état infecté est globalement stable.")
    else:
        print("❌ Condition non satisfaite : l’état infecté converge vers (0,lambda/gamma,O).")

# Cas 1 : condition satisfaite
check_infection_stability(tau=1.5, lam=2, gamma=1, x0=0, beta=0.5, mu=1)

# Cas 2 : condition non satisfaite
check_infection_stability(tau=1, lam=1, gamma=2.0, x0=1.0, beta=0.1, mu=1.0)

```

## 3. Distribution en tailles des agrégats

Jusqu’à présent, nous avons étudié des quantités globales comme le nombre d’agrégats $U(t)$ ou la masse totale $P(t)$.  
Cependant, ces agrégats peuvent avoir des tailles variées, et l’évolution de leur **distribution en taille** est un aspect crucial pour comprendre la dynamique du prion.

Nous nous intéressons ici à la **densité stationnaire** $u_\infty(x)$ des agrégats de taille $x$, lorsque le système atteint un régime infecté stable.  
On suppose que cette densité est **indépendante du temps** et on cherche à déterminer sa forme.

---

### Équation de distribution stationnaire

À partir du système complet en dimension continue (équation (2) du texte), on étudie les solutions stationnaires $(V_\infty, u_\infty(x))$ qui satisfont le système :

$
\begin{cases}
V_\infty \left( \gamma + \displaystyle\int_{x_0}^\infty \tau(x) u_\infty(x) \, dx \right) = \lambda + 2 \displaystyle\int_0^{x_0} x \left( \int_{x_0}^\infty \bar{\beta} y \kappa(x,y) u_\infty(y) \, dy \right) dx \\
\\
V_\infty \dfrac{d}{dx} \left( \tau(x) u_\infty(x) \right) = -(\mu + \bar{\beta} x) u_\infty(x) + 2 \displaystyle\int_x^\infty \bar{\beta} y \kappa(x, y) u_\infty(y) \, dy
\end{cases}
$

où :
- $\tau(x)$ est le taux de polymérisation en fonction de la taille,
- $\bar{\beta} x$ est le taux de fragmentation,
- $x_0$ est la **taille critique minimale**, en dessous de laquelle les agrégats sont instables.

---

### Simplifications et hypothèses

Afin de rendre ce système traitable numériquement, on adopte les hypothèses suivantes :

- $\mu$ constant,
- $\beta(x) = \bar{\beta} x$,
- $\kappa(x,y) = \dfrac{1}{y}$ si $x < y$ et $y > x_0$, et $0$ sinon.

---

### Déduction du schéma stationnaire

Sous ces hypothèses, et en développant les intégrales, la deuxième équation devient une équation différentielle linéaire en $x$ :

$
V_\infty \, \dfrac{d^2}{dx^2} (\tau(x) u_\infty(x)) + \dfrac{d}{dx} \left( (\mu + \bar{\beta} x) u_\infty(x) \right) + 2 \bar{\beta} u_\infty(x) = 0
$

avec les conditions aux bords :

$
u_\infty(x_0) = 0, \quad V_\infty \frac{d}{dx}(\tau u)(x_0) = 2 \bar{\beta} \int_{x_0}^{\infty} u(y) \, dy
$

Ce modèle décrit comment la distribution en tailles $u_\infty(x)$ se stabilise dans le temps, selon le profil de $\tau(x)$.

Nous allons maintenant mettre en œuvre un **schéma numérique** pour approximer cette solution $u_\infty(x)$ :


### Schéma numérique pour la distribution stationnaire des agrégats


On discrétise l’intervalle $[x_0, x_{\max}]$ en $N$ points réguliers $x_i = x_0 + i h$, avec $h$ le pas de discrétisation. L'équation différentielle devient un **système linéaire tridiagonal** pour les valeurs $u_i \approx u_\infty(x_i)$.

Le texte (équation (13)) propose une discrétisation directe de l’équation stationnaire, qui s’écrit pour tout $i \ge 1$ :

$$
\frac{1}{h^2} \left( \tau_{i+1} u_{i+1} - 2 \tau_i u_i + \tau_{i-1} u_{i-1} \right)
+ \frac{1}{2 h V} \left( (\mu + \bar{\beta} x_{i+1}) u_{i+1} - (\mu + \bar{\beta} x_{i-1}) u_{i-1} \right)
+ \frac{2 \bar{\beta}}{V} u_i = 0
$$

Ce schéma s’écrit de manière explicite en isolant $u_{i+1}$ :

$
u_{i+1} = \frac{-B_i - C_i - D_i}{A_i}
$

avec les coefficients définis comme suit :

- $A_i = \dfrac{\tau_{i+1}}{h^2} + \dfrac{\mu + \bar{\beta} x_{i+1}}{2 h V}$  
- $B_i = \dfrac{-2 \tau_i u_i + \tau_{i-1} u_{i-1}}{h^2}$  
- $C_i = -\dfrac{(\mu + \bar{\beta} x_{i-1}) u_{i-1}}{2 h V}$  
- $D_i = \dfrac{2 \bar{\beta}}{V} u_i$

On commence l’itération avec la condition $u_0 = 0$ et une **condition de Neumann** donnée par la dérivée :

$
\frac{u_1 - u_0}{h} = \frac{2 \bar{\beta}}{V}
\quad \Rightarrow \quad u_1 = \frac{2 \bar{\beta}}{V} h
$

#### Algorithme de point fixe

Le paramètre $V$ est mis à jour à chaque itération à l’aide d’une **formule intégrée** dérivée de l’équation de conservation (équation (16)) :

$
V = \frac{ \mu \int x u(x) dx + \bar{\beta} x_0^2 \int u(x) dx }{ \int \tau(x) u(x) dx }
$

On normalise la solution $u$ pour imposer $\int u(x) dx = 1$, ce qui revient à fixer une masse totale unitaire. On répète ce processus jusqu’à convergence du couple $(V, u)$.

---

Ce schéma permet de générer numériquement des profils stationnaires que nous allons comparer dans les cas suivants :
- $\tau(x)$ **constant** : profil simple, unimodal,
- $\tau(x)$ **variable** : apparition de **profils bimodaux** plus riches, en accord avec les observations biologiques.


```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
x0 = 0  # Taille critique
beta_bar = 0.5  # Coefficient de fragmentation
mu = 1  # Taux de dégradation
tau0 = 0.001  # Partie constante de τ(x)
A = 0.1  # Amplitude de la variation de τ(x)
m = 2  # Centre de la gaussienne dans τ(x)
sigma = np.sqrt(0.1)  # Écart-type dans τ(x)

# Discrétisation
h = 0.01  # Pas d'espace
x_max = 10  # Domaine spatial
N = int(x_max / h)  # Nombre de points
x_values = np.linspace(x0, x_max, N)  # Points de discrétisation

# Fonction τ(x) variable
def tau(x):
    return tau0 + A * np.exp(-((x - m) ** 2) / sigma**2)

def tau_constant(x):
    return 1

def main(tau, V0 = 2,tol = 1e-6, max_iter = 100):
    V = 0  
    V_new = V0
    iter_num = 1
    while (iter_num < max_iter) and (np.abs(V_new - V) > tol):
        V = V_new
        u = np.zeros(N)
        num_i = 0
        denom_i = 0
        # Conditions aux limites (équation 14)
        u[0] = 0  # u∞(x0) = 0
        u[1] = (2 * beta_bar / V) * h  # Condition dérivée avec prévention de division par zéro
    
        for i in range(1, N - 1):
            tau_i = tau(x_values[i])
            tau_ip1 = tau(x_values[i + 1])
            tau_im1 = tau(x_values[i - 1])
        
            A_i = tau_ip1/h**2 + (mu+beta_bar*x_values[i + 1])/(2*h*V)
            B_i = (-2 * tau_i * u[i] + tau_im1 * u[i - 1])/h**2
            C_i = -(((mu + beta_bar * x_values[i - 1]) * u[i - 1]) / (2 * h * V))
            D_i = (2 * beta_bar / V) * u[i]
        
            u[i+1] = (-B_i - C_i - D_i)/A_i
        
            num_i += i * h * u[i]
            denom_i += tau_i * u[i]

        # Mise à jour de V
        V_new = (mu * h * num_i + beta_bar * x0**2) / (h * denom_i)
        iter_num += 1
    
    print("Convergence atteinte en :", iter_num, "itérations")
    return u

u_tau = main(tau)
u_constant = main(tau_constant)

# Affichage des résultats
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Graphique 1 : tau(x) variable
ax1.plot(x_values, u_tau, label=r"Solution avec $\tau(x)$ variable", color='purple')
ax1.set_xlabel("x")
ax1.set_ylabel("Prion (PrPSc)")
ax1.set_title(r"Distribution avec $\tau(x)$ variable")
ax1.legend()
ax1.grid()

# Graphique 2 : tau(x) constant
ax2.plot(x_values, u_constant, label=r"Solution avec $\tau(x) = 1$", color='orange')
ax2.set_xlabel("x")
ax2.set_ylabel("Prion (PrPSc)")
ax2.set_title(r"Distribution avec $\tau(x) = 1$")
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()

```

### Analyse des résultats numériques

La figure ci-dessous présente les solutions stationnaires $u_\infty(x)$ obtenues pour deux cas :

- **à gauche** : profil obtenu avec un taux de polymérisation $\tau(x)$ **variable**, de type Gaussien centré autour d’une taille privilégiée,
- **à droite** : profil obtenu avec un taux $\tau(x) \equiv 1$, **constant** sur tout le domaine.

#### Cas $\tau(x) \equiv 1$ — profil unimodal

Lorsque $\tau(x)$ est constant, on observe une **distribution unimodale** : la densité des agrégats présente un pic unique centré autour d'une taille intermédiaire.  
Cela correspond à une croissance homogène et indifférenciée des agrégats, sans préférence pour une taille spécifique. La fragmentation agit de manière régulière, ce qui conduit à une concentration de masse autour d’une valeur moyenne, puis une décroissance progressive vers les grandes tailles.

#### Cas $\tau(x)$ variable — profil bimodal

Lorsque $\tau(x)$ varie avec la taille (par exemple un profil Gaussien centré), on observe une **distribution bimodale** : deux pics nets apparaissent, correspondant à deux tailles d’agrégats préférentiellement formés.

Ce comportement s’explique biologiquement par le fait que certaines tailles d’agrégats sont **favorisées cinétiquement** : la polymérisation est plus efficace autour d’une taille donnée, tandis que les très petits ou très grands agrégats sont moins actifs. Cela reflète les données expérimentales, qui montrent souvent des distributions en tailles non triviales dans les maladies à prions.

---

### 🧠 Interprétation biologique

- Le modèle montre que la structure du taux de polymérisation $\tau(x)$ influence directement la **forme de la distribution stationnaire**.
- L’apparition de **deux populations préférentielles** d’agrégats peut expliquer certaines observations biologiques dans lesquelles coexistent petits noyaux (graines) et agrégats plus larges.
- Ces résultats montrent l’importance de modéliser **finement les mécanismes de croissance** pour rendre compte des phénomènes infectieux observés in vivo.
