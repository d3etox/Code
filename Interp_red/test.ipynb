```python
import numpy as np

def point_in_triangle_3d(P, A, B, C, tol=1e-9):
    """
    Vérifie si le point P est dans le triangle défini par (A, B, C) en 3D.
    
    Paramètres :
    -----------
    P, A, B, C : array-like de taille 3
        Coordonnées des points (x, y, z).
    tol : float
        Tolérance numérique pour la coplanarité.
    
    Retour :
    --------
    True si P est dans le triangle, False sinon.
    """
    # Conversion en numpy
    P, A, B, C = map(np.array, (P, A, B, C))
    
    # Vecteurs
    v0 = C - A
    v1 = B - A
    v2 = P - A

    # Vérification coplanarité : (P-A) doit être orthogonal à n = (B-A) x (C-A)
    n = np.cross(B - A, C - A)
    if abs(np.dot(n, P - A)) > tol:
        return False  # P n'est pas dans le plan

    # Produits scalaires
    d00 = np.dot(v0, v0)
    d01 = np.dot(v0, v1)
    d11 = np.dot(v1, v1)
    d20 = np.dot(v2, v0)
    d21 = np.dot(v2, v1)

    # Déterminant
    denom = d00 * d11 - d01 * d01
    if abs(denom) < tol:
        return False  # Triangle dégénéré

    # Coordonnées barycentriques
    beta = (d11 * d20 - d01 * d21) / denom
    gamma = (d00 * d21 - d01 * d20) / denom
    alpha = 1 - beta - gamma

    # Test appartenance (dans [0,1] avec tolérance)
    return (alpha >= -tol) and (beta >= -tol) and (gamma >= -tol)

def triangle_area(A, B, C):
    """
    Calcule l'aire d'un triangle en 3D à partir des sommets A, B, C.
    
    Paramètres :
    -----------
    A, B, C : array-like (x, y, z)
    
    Retour :
    --------
    float : aire du triangle
    """
    A, B, C = map(np.array, (A, B, C))
    return 0.5 * np.linalg.norm(np.cross(B - A, C - A))

# Exemple d'utilisation
A = [0, 0, 0]
B = [0, 1, 0]
C = [1, 0, 0]
X1 = [0.51, 0.51, 0]
X2 = [0.5, 0.51, 0]
print(point_in_triangle_3d(X1, A, B, C))
print(point_in_triangle_3d(X2, A, B, C)) 
print(triangle_area(A, B, C))
```
